```python
import numpy as np
import math
import matplotlib.pyplot as plt
x5 = np.arange(0, math.pi*2, 0.05)
y5 = np.sin(x5)
plt.xlim(0,10)
plt.ylim(0,1.5)
plt.xticks([0,2,4,6,8,10])
plt.yticks([-1,0,1])
plt.plot(x5,y5)
plt.show()
```

```python
fig = plt.figure(figsize=(8,5))
a5 = fig.add_axes([0.1,0.1,0.8,0.8])
a5.plot(x5,np.exp(x5))
a5.set_ylabel('exp')
a6 = a5.twinx() # 축이 또 하나 생김
a6.plot(x5,np.log(x5),'rx')
a6.set_ylabel('log')
fig.legend(labels=('exp','log'),loc='upper left')
plt.show()
```
