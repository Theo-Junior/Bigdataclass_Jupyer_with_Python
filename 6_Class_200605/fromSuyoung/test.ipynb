```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

x = np.arange(0, math.pi * 2, 0.05)
y = np.sin(x)
plt.plot(x, y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Sine Graph")
plt.show()
```

```python
from pylab import *
x = linspace(-3,3,30)
y = x**2 #y=x제곱
y1 = x**4 #y=x네제곱
plot(x,y,'r')
plot(x,y1,'b')
plot(x,y1+3,'m')
# symbols : -,+....
# b,g,r,c,m,y,k,w
show()
```

```python
plot(x, sin(x))
plot(x, cos(x), 'r.')
plot(x, -sin(x), 'g+')
show()
```

```python
x1 = np.arange(0, math.pi*2, 0.05)
y1 = np.sin(x1)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1]) # left, bottom, width, height
ax.plot(x1,y1)
ax.legend(labels=('x'), loc='lower right')
ax.set_title('Sine grah')
ax.set_xlabel("x axis")
ax.set_ylabel("y axis")
plt.show()
```

```python
x3 = np.arange(0, math.pi * 2, 0.05)
y3 = sin(x3)
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.55,0.55,0.3,0.3])
axes1.plot(x3, y3, 'b')
axes2.plot(x3, np.cos(y3),'r')
axes1.set_title('sine')
axes2.set_title('cosine')
plt.show()

```
