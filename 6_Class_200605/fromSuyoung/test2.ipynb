```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
fig = plt.figure()
x = np.arange(1,5)
a = fig.subplots(2,2)
a[0][0].plot(x, x*x, 'red', lw=10)
a[0][0].grid(True) # 표출력 유무
a[0][0].set_title("square") 
a[0][1].plot(x, np.sqrt(x))
a[0][1].grid(color='blue', ls='-.', lw=1) # ls선종류, lw선의 두께
a[0][1].set_title("square root")
a[1][0].plot(x, np.exp(x))
a[1][1].plot(x, np.log10(x))
plt.show()

```

```python
fig = plt.figure(figsize=(4,3))
a1 = fig.add_axes([0,0,1,1])
x4 = np.arange(1,10)
a1.plot(x4, np.exp(x4))
a1.set_title('exp')
a1.set_ylim(0,10000) # y축의 표시범위
a1.set_xlim(0,10) # x축의 표시범위
plt.show()
```

```python
# from pylab import *
import numpy as np
import math
import matplotlib.pyplot as plt
x5 = np.arange(0, math.pi*2, 0.05)
y5 = np.sin(x5)
plt.xlim(0,10)
plt.ylim(0,1.5)
plt.xticks([0,2,4,6,8,10])
plt.yticks([-1,0,1])
plt.plot(x5,y5)
plt.show()
```

```python
fig = plt.figure(figsize=(8,5))
a5 = fig.add_axes([0.1,0.1,0.8,0.8])
a5.plot(x5,np.exp(x5))
a5.set_ylabel('exp')
a6 = a5.twinx() #x축은 그대로 놓고, y축만 다시 정하는 것
a6.plot(x5,np.log(x5),'rx')
a6.set_ylabel('log')
fig.legend(labels=('exp','log'), loc='upper left')
plt.show()
```
