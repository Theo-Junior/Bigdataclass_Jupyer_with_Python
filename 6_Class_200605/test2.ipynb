```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

fig = plt.figure()
x = np.arange(1,5)
a = fig.subplots(2,2)

a[0][0].plot(x, x*x)
a[0][0].set_title("square")
a[0][1].plot(x, np.sqrt(x))
a[0][1].set_title("square root")
a[1][0].plot(x, np.exp(x))
a[1][1].plot(x, np.log10(x))
plt.show()
```

```python
fig = plt.figure(figsize=(4,3))
a1 = fig.add_axes([0,0,1,1])
x4 = np.arange(1,10)
a1.plot(x4, np.exp(x4),'red')
a1.set_title("exp")
# 그냥 flot으로 출력할 때는 xlim, ylim만 작성
a1.set_ylim(0,10000)
a1.set_xlim(0,20)
plt.show()
```
