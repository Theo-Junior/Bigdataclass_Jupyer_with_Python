```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel('서울시 비만도 통계.xlsx',encoding='cp949',header=2,index_col= 1)
data = data.drop(['기간'],axis=1)
data = data.iloc[1:,:]
a = data['비만(BMI≥25)'].values
b = data['저체중(BMI<18.5)'].values
c = data['정상(18.5≤BMI＜25)'].values
t = np.arange(25)
idx = data.index.values
fig = plt.figure(figsize=(10,3))
ax6 = fig.add_axes([0,0,1,1])
ax6.bar(t+0.00,a,width=0.2)
ax6.bar(t+0.25,b,width=0.2)
ax6.bar(t+0.5,c,width=0.2)
plt.xticks(t,idx,rotation = 45)
fig.legend(labels=('비만','저체중','정상'),loc='best')
```
